```python
import pandas as pd

movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")

movies = movies.merge(credits, on='title')
movies
```

```python
movies = movies[['movie_id', 'original_language', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies
```

```python
movies['genres']
```

```python
import ast

def convert(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l

def convert_cast(text):
    l = []
    count = 0
    
    for i in ast.literal_eval(text):
        if count < 3:
            l.append(i['name'])
    return l

def director(text):
    l = []
    
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['cast'] = movies['cast'].apply(convert_cast)
movies['crew'] = movies['crew'].apply(director)

movies

```

```python
def extract_first_4_names(cast_list):
    return cast_list[:4] 

movies['cast'] = movies['cast'].apply(extract_first_4_names)

movies['cast']
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('punkt', download_dir='/Users/viditparashar/nltk_data')
nltk.download('stopwords', download_dir='/Users/viditparashar/nltk_data')
nltk.download('wordnet', download_dir='/Users/viditparashar/nltk_data')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def pre_process_text(text):
    if isinstance(text, str):
        text = text.lower()
        words = word_tokenize(text)
        words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words and word.isalpha()]
        return ' '.join(words)
    return ''

movies['overview'] = movies['overview'].apply(pre_process_text)

print(movies['overview'])
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
print(movies['overview'])
```

```python
def remove_space(word):
    l = []
    for i in word:
        l.append(i.replace(" ",""))
    return l    
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)

movies['cast']
movies['crew']
```

```python
movies
```

```python
movies['tags'] = movies['overview'] + movies['cast'] + movies['crew'] + movies['genres'] + movies['keywords']

movies['tags']
```

```python
movies = movies.drop(columns = ['overview', 'cast', 'crew', 'genres', 'keywords'])
movies
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))

movies['tags']
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

def stem_words(word):
    l = []
    for i in word.split():
        l.append(ps.stem(i))
    return ' '.join(l)
```

```python
movies['tags'] = movies['tags'].apply(stem_words)

movies['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000)

vector = cv.fit_transform(movies['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vector)
```

```python
movies
```

```python
def recommend(movie):
    index = movies[movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])), key=lambda x: x[1], reverse=True)
    
    for i in distances[1:6]:
        print(movies.iloc[i[0]].title)
```

```python
recommend('The Dark Knight Rises')
```
